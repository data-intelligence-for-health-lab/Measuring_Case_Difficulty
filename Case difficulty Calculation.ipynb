```python
def two_class_equation_comparison(df):
  print("two_class_equation_working")
  equation2_difficulties = []
  #print(len(df.iloc[0,4:])) #need to be 100, means 100 predictions made

  # Equation 2
  def location(i):
    return abs(statistics.mean(df.iloc[i,4:])-df.iloc[i,3])
  def distribution(i):
    return statistics.stdev(df.iloc[i,4:])/math.sqrt(1/12)

  euqation2_row = []
  for i in range(len(df)):
    difficulty = (location(i) + distribution(i))/2
    euqation2_row.append(difficulty)
  equation2_difficulties.append(euqation2_row)

  new_df = df.iloc[:,1:4]
  new_df = new_df.rename({1: 'x1', 2: 'x2', 3:'label'}, axis=1)
  new_df['difficulty'] = equation2_difficulties[0]
  return(new_df,equation2_difficulties)


def three_class_equation_comparison(grouped_columns,original_data):
  print("three_class_equation_working")
  equation2_difficulties = []

  df = original_data
  ground_truth = original_data[3]

  # Equation 2 only need ground truth
  def location(i):
    if ground_truth[i] == 0:
      locaion_0 = abs(statistics.mean(grouped_columns[0].iloc[i,:])-1)
      locaion_1 =abs(statistics.mean(grouped_columns[1].iloc[i,:])-0)
      locaion_2 =abs(statistics.mean(grouped_columns[2].iloc[i,:])-0)
    elif ground_truth[i] == 1:
      locaion_0 = abs(statistics.mean(grouped_columns[0].iloc[i,:])-0)
      locaion_1 =abs(statistics.mean(grouped_columns[1].iloc[i,:])-1)
      locaion_2 =abs(statistics.mean(grouped_columns[2].iloc[i,:])-0)
    elif ground_truth[i] == 2:
      locaion_0 = abs(statistics.mean(grouped_columns[0].iloc[i,:])-0)
      locaion_1 =abs(statistics.mean(grouped_columns[1].iloc[i,:])-0)
      locaion_2 =abs(statistics.mean(grouped_columns[2].iloc[i,:])-1)
    return locaion_0,locaion_1,locaion_2

  def distribution(i):
    distribution_0 = statistics.stdev(grouped_columns[0].iloc[i,:])/math.sqrt(1/12)
    distribution_1 = statistics.stdev(grouped_columns[1].iloc[i,:])/math.sqrt(1/12)
    distribution_2 = statistics.stdev(grouped_columns[2].iloc[i,:])/math.sqrt(1/12)
    return distribution_0,distribution_1,distribution_2

  # Equation 2
  euqation2_row = []
  for i in range(len(ground_truth)):
    #print(len(grouped_columns[0].iloc[i,:])) #need to be 100, means 100 predictions made
    locaion_0,locaion_1,locaion_2 = location(i)
    distribution_0,distribution_1,distribution_2 = distribution(i)
    first_class_difficulty = (locaion_0 + distribution_0)/2
    second_class_difficulty = (locaion_1 + distribution_1)/2
    third_class_difficulty = (locaion_2 + distribution_2)/2
    difficulty = statistics.mean([first_class_difficulty,second_class_difficulty,third_class_difficulty])
    euqation2_row.append(difficulty)

  equation2_difficulties.append(euqation2_row)

  new_df = df.iloc[:,1:4]
  new_df = new_df.rename({1: 'x1', 2: 'x2', 3:'label'}, axis=1)
  new_df['difficulty'] = equation2_difficulties[0]
  return(new_df,equation2_difficulties)
```

```python
def data_split(df):
  #Splitting data
  ground_truth = 3
  column_range = df.iloc[:, ground_truth+1:304]
  grouped_columns = []
  for i in range(ground_truth):
      group = column_range.iloc[:, i::df[ground_truth].nunique()]  # Selecting columns with an offset of multiclass number, 3 is the label column
      grouped_columns.append(group)
  #Sending original data
  original_data = df.iloc[:,0:ground_truth+1]
  return(grouped_columns,original_data)
```

```python
app3_difficulty = []
simulated_app3 = []

for filename in file_list3:
  with open(os.path.join(os.getcwd(), filename), 'r') as f: # open in read only mode
    df = pd.read_excel(filename, header = None)
    if len(df.columns) == 104:
      print(filename,'is two class')
      temp_data, eq2 = two_class_equation_comparison(df)
    if len(df.columns) == 304:
      print(filename,'is three class')
      grouped_columns,original_data = data_split(df)
      temp_data, eq2 = three_class_equation_comparison(grouped_columns,original_data)

    simulated_app3.append(temp_data)
    app3_difficulty.extend(eq2)
```

```python
import pandas as pd
import numpy as np
from scipy import stats

from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

import statistics
import math

from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
# Data Preprocessing
# Define the column types
categorical_columns = ['Gender', 'Ever_Married', 'Graduated','Profession','Spending_Score','Var_1']
numeric_columns = ['Age', 'Work_Experience', 'Family_Size']

# categorical columns (imputation, one hot encoder)
# numeric columns (imputation, standardscaler)
categorical_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('one_hot_encoder', OneHotEncoder(handle_unknown='ignore'))
])

numeric_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('scaler', StandardScaler())
])

def one_hot_encoding(original_data):
  # Customer Data preprocessing
  categorical_transformed = pd.get_dummies(original_data[categorical_columns], columns=categorical_columns)
  numeric_transformed = pd.DataFrame((original_data[numeric_columns]), columns=numeric_columns)

  # Concatenate the transformed columns back into a new DataFrame
  one_hot_data = pd.concat([categorical_transformed, numeric_transformed, original_data.drop(categorical_columns + numeric_columns, axis=1)], axis=1)
  one_hot_data
  return(one_hot_data)


def standard(original_data):
  # Customer Data preprocessing
  categorical_transformed = original_data.drop(columns=numeric_columns)
  numeric_transformed = pd.DataFrame(numeric_transformer.fit_transform(original_data[numeric_columns]), columns=numeric_columns)

  # Concatenate the transformed columns back into a new DataFrame
  standard_data = pd.concat([numeric_transformed, categorical_transformed], axis=1)
  standard_data
  return(standard_data)
```

```python
def three_class_equation_comparison(grouped_columns,original_data):
  ground_truth = original_data[10]
  print(ground_truth)

  # Equation 2 only need ground truth
  def location(i):
    if ground_truth[i] == 0:
      locaion_0 = abs(statistics.mean(grouped_columns[0].iloc[i,:])-1)
      locaion_1 =abs(statistics.mean(grouped_columns[1].iloc[i,:])-0)
      locaion_2 =abs(statistics.mean(grouped_columns[2].iloc[i,:])-0)
      locaion_3 =abs(statistics.mean(grouped_columns[3].iloc[i,:])-0)
    elif ground_truth[i] == 1:
      locaion_0 = abs(statistics.mean(grouped_columns[0].iloc[i,:])-0)
      locaion_1 =abs(statistics.mean(grouped_columns[1].iloc[i,:])-1)
      locaion_2 =abs(statistics.mean(grouped_columns[2].iloc[i,:])-0)
      locaion_3 =abs(statistics.mean(grouped_columns[3].iloc[i,:])-0)
    elif ground_truth[i] == 2:
      locaion_0 = abs(statistics.mean(grouped_columns[0].iloc[i,:])-0)
      locaion_1 =abs(statistics.mean(grouped_columns[1].iloc[i,:])-0)
      locaion_2 =abs(statistics.mean(grouped_columns[2].iloc[i,:])-1)
      locaion_3 =abs(statistics.mean(grouped_columns[3].iloc[i,:])-0)
    elif ground_truth[i] == 3:
      locaion_0 = abs(statistics.mean(grouped_columns[0].iloc[i,:])-0)
      locaion_1 =abs(statistics.mean(grouped_columns[1].iloc[i,:])-0)
      locaion_2 =abs(statistics.mean(grouped_columns[2].iloc[i,:])-0)
      locaion_3 =abs(statistics.mean(grouped_columns[3].iloc[i,:])-1)

    return locaion_0,locaion_1,locaion_2,locaion_3

  def distribution(i):
    distribution_0 = statistics.stdev(grouped_columns[0].iloc[i,:])/math.sqrt(1/12)
    distribution_1 = statistics.stdev(grouped_columns[1].iloc[i,:])/math.sqrt(1/12)
    distribution_2 = statistics.stdev(grouped_columns[2].iloc[i,:])/math.sqrt(1/12)
    distribution_3 = statistics.stdev(grouped_columns[3].iloc[i,:])/math.sqrt(1/12)
    return distribution_0,distribution_1,distribution_2,distribution_3

  # Equation 2
  euqation2_row = []
  for i in range(len(ground_truth)):
    locaion_0,locaion_1,locaion_2,locaion_3 = location(i)
    distribution_0,distribution_1,distribution_2,distribution_3 = distribution(i)
    first_class_difficulty = (locaion_0 + distribution_0)/2
    second_class_difficulty = (locaion_1 + distribution_1)/2
    third_class_difficulty = (locaion_2 + distribution_2)/2
    fourth_class_difficulty = (locaion_3 + distribution_3)/2
    difficulty = statistics.mean([first_class_difficulty,second_class_difficulty,third_class_difficulty,fourth_class_difficulty])
    euqation2_row.append(difficulty)

  new_df = original_data
  new_df['difficulty'] = euqation2_row
  new_df.columns = ['Gender', 'Ever_Married', 'Age', 'Graduated', 'Profession',
       'Work_Experience', 'Spending_Score', 'Family_Size', 'Var_1', 'label',
       'difficulty']
  return(new_df,euqation2_row)
```
